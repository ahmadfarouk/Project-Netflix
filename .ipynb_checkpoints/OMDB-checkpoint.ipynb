```python
# Dependencies
import pandas as pd
import requests
from config import api_key
from pprint import pprint

#file path to open CSV
movie_file ="Resources/netflix_titles.csv"
```

```python
# Read and display the CSV with Pandas

movie_file_df = pd.read_csv(movie_file)
movie_file_df.head()
```

```python
base_url = f"http://www.omdbapi.com/?apikey={api_key}&t="
```

```python
titles= movie_file_df["title"]
Genres = []
awards = []
released_date = []
imdb_ratings = []
rt_ratings = []

#print(titles)
i = 0
for title in titles:
    i= i+1
    #if i <=16:
    url = base_url +title   
    movie = requests.get(url).json()
        #pprint(movie)

    try:
        Genres.append(movie["Genre"])
    except:
        Genres.append("NaN")
    try:
         awards.append(movie["Awards"])
    except:
        awards.append("NaN")   
    try:
        released_date.append(movie["Released"])
    except:
        released_date.append("NaN") 
    try:
        imdb_ratings.append(movie["Ratings"][0]["Value"])
    except:
        imdb_ratings.append("0")      

    try:
        rt_ratings.append(movie["Ratings"][1]["Value"])
    except:
        rt_ratings.append("0")
    
       
        
print (Genres)
print(awards)
print(released_date)
print(imdb_ratings)
print(rt_ratings)
```

```python
 #add the information extracted to the DataFrame
print(len(imdb_ratings))
movie_file_df["IMDB rating"] = imdb_ratings
movie_file_df["Rotten Tomatoes rating"] = rt_ratings
movie_file_df["Award"] = awards
movie_file_df["Released Date"] = released_date


```
